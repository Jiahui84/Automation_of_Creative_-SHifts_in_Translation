```python
#open file and turn it into a list with each sentence as an element in a list
f = open("NoUCP.txt",encoding = "utf-8")
lines = f.readlines()

lines
```

```python
#test if every line is an element
lines[0]
```

```python
#count the length of cery line in the list and append them in a new list
list= []
for i in lines:
    length = len(i)
    list.append(length)

list
```

```python
#count the average length of the list
mean(list)
```

```python
#see is there's extreme value in the list 
max(list)
```

```python
#see is there's extreme value in the list 
min(list)
```

```python
# set a new function to count the average length of the list without extreme values
def my_AVERAGE_main(data_list):
    
    if len(data_list)==0:
        return 0
    if len(data_list)>2:
        data_list.remove(min(data_list))
        data_list.remove(max(data_list))
        average_data = float(sum(data_list))/len(data_list)
        return average_data
    elif len(data_list)<=2:
        average_data = float(sum(data_list))/len(data_list)
        return average_data
```

```python
average = my_AVERAGE_main(list)
print(average)
```
