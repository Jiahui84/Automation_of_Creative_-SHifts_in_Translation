```python
#import the needed library
from gensim.models import FastText
import numpy as np
```

```python
#load the English wiki model
FASTTEXTFILE_EN = "wiki.en.bin"
FASTTEXT_EN=FastText.load_fasttext_format(FASTTEXTFILE_EN)
```

```python
#check if the model can query vectors
FASTTEXT_EN.wv['Jackson'].tolist()
```

```python
#query the vectors of every word in English txt in the wiki English model, and write the word and its vector in a txt file
fw = open('EN_EMB.txt', 'w', encoding='utf-8')
with open('.txt', encoding='utf-8') as fr:
    for line in fr:
        for l in line.strip().split():
            try:
                wc = FASTTEXT_EN.wv[l.strip().strip('\ufeff')].tolist()
                fw.write(l+' '+' '.join(list(map(str,wc))) + '\n')
                fw.flush()
                print(l)
            except:
                wc = list(np.random.rand(300))
                fw.write(l+' '+' '.join(list(map(str,wc))) + '\n')
                fw.flush()
```
