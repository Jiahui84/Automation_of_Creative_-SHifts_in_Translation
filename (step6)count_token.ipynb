```python
#tokenize the text by space with every word as a token
list = []
with open('model.txt', encoding='utf-8') as fr:
    for line in fr:
        for l in line.strip().split():
            list.append(l)
```

```python
#check the number of the tokens; the number and dimension will be written in the first line EMB file according to the input format of vecmap
len(list)
```
