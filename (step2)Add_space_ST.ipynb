```python
#import the tokenizationtool
from sacremoses import MosesTokenizer, MosesDetokenizer
```

```python
#set the language as English
tokenizer = MosesTokenizer('en')
```

```python
#read the English ST
ST =  open ("desktop/ST.txt", "r")
ST_new = ST.readlines()
ST_new
```

```python
#import html for decoding html entities 
import html
```

```python
#tokenize the text into words per sentence
list = []
for i in ST_new:
    space = tokenizer.tokenize(i)
    list.append(space)
```

```python
list
```

```python
#write the results in the txt
with open ('写入.txt', 'w') as output:
    for i in list:
        for word in i:
            output.write(str(html.unescape(word))+' ')
            output.write(' ')
        output.write('\n')
```
