```python
#import the needed libraries
import jieba
import pandas as pd
import string
import numpy as np
import gensim
from gensim.models import KeyedVectors
import os
import matplotlib.pyplot as plt
import nltk
```

```python
#read the ST
txt = open("newstest2020.en-zh.Huawei_TSC.618.txt", "r", encoding='UTF-8-sig').read()
```

```python
#make comparison of three tokenization modes
# cut words with paddle mode
words = jieba.lcut(txt) 
words
```

```python
#output the processed text
fW = open('MT.txt', 'w', encoding='UTF-8-sig')
fW.write(' '.join(words))
```

```python
# cut words with full mode
words_full = jieba.lcut(txt,cut_all=True) 
words_full
```

```python
#cut words with search mode
words_search = jieba.lcut_for_search(txt) 
words_search
```
