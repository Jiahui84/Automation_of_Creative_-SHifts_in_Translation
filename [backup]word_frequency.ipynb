```python
#read the text file
f = open("NoUCP.txt",encoding = "utf-8")
lines = f.read()

lines
```

```python
#lowercase all the charaters
new_text = lines.lower()
new_text
```

```python
#set a function to remove punctuations
def remove_punc(text):
    punctuation = '!@#$%^&*()_-+={}[]:;"\'|<>,.?/~`\n\r'
    return ''.join(character for character in text
                   if character not in punctuation)
```

```python
#get the text without punctuations
no_pun = remove_punc(new_text)
no_pun
```

```python
#import needed libriaries for romving stops words in the text
import nltk
nltk.download('stopwords')
```

```python
#tokenize the text
tokens = nltk.word_tokenize(no_pun)
tokens
```

```python
#remoce all the stopwords in the text
from nltk.corpus import stopwords
en_stops = set(stopwords.words('english'))
list = []
for word in tokens: 
    if word not in en_stops:
        list.append(word)
```

```python
#get the word list without stopwords
list
```

```python
#import the needed libraries
import codecs
from pylab import mpl

mpl.rcParams['font.sans-serif'] = ['Calibri']  # Specify the default font
mpl.rcParams['axes.unicode_minus'] = False  # Solve the problem that the minus sign'-' is displayed as a square in the saved image

word = []#set new word list for unique words
counter = {}#count the words

for i in list:
    if len(i) == 0:#in case there are any blank prevent the code from running
        continue
    if not i in word:#every word is only written once in the list
        word.append(i)
    if not i in counter:#count how many times the word shows up 
        counter[i] = 0
    else:
        counter[i] += 1

counter_list = sorted(counter.items(), key=lambda x: x[1], reverse=True)#set the order of presenting, so as to get the most frequent words 

print(counter_list[:20])
```
