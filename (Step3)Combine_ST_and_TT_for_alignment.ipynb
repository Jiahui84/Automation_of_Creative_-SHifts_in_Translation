```python
#open three files: source text target text of human translators and target text of machine translation
file1 = open("Space_ST.txt",'r', encoding='UTF-8-sig')
file2 = open("MT.txt",'r', encoding='UTF-8-sig')
file3 = open("HT.txt",'r', encoding='UTF-8-sig')
```

```python
#split the texts into lists of sentences
file1_lists = file1.read().splitlines()
file2_lists =file2.readlines()
file3_lists =file3.readlines()
```

```python
#check what is in the list
file1_lists
```

```python
#check what is in the list
file2_lists
```

```python
#check what is in the list
file3_lists
```

```python
#combine ST and MT with "|||" as separator, and put the preprocessed text in a list
main_file_MT = []

for i in range(len(file1_lists)):
    s = ""
    s = "".join([file1_lists[i]," |||",file2_lists[i]])
    main_file_MT.append(s)
```

```python
main_file_MT
```

```python
#write every line in a list
file4 = open("ST-MT_for_input.txt","w",encoding='UTF-8-sig')

file4.writelines(main_file_MT)
```
